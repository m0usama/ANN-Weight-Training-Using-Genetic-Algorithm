```python
import pandas as pd
import random
import math
```

```python
dataset = pd.read_csv('iris_dataset_training.csv')
row = dataset.shape[0]
col = dataset.shape[1]
hiddenNodes=8

class weightVector:

  def __init__(self):
    self.weights1 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      arr = []
      for j in range(0,col-1): 
        arr.append(float(random.uniform(-1.0,1.0)))
      self.weights1.append(arr)
    self.bias1 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      inp = 0
      while(inp == 0):
        inp = float(random.uniform(-1.0,1.0))
        self.bias1.append(inp)
    self.weights2 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      self.weights2.append(float(random.uniform(-1.0,1.0)))
    inp = 0
    while(inp == 0):
      inp = float(random.uniform(-1.0,1.0))
      self.bias2 = inp
    

    fitness=0

  def setContents(self,gene):
    index=0
    self.weights1=[]
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      arr = []
      for j in range(0,col-1): #noumber of inputs
        arr.append(gene[index])
        index+=1
      self.weights1.append(arr)
    self.bias1 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      self.bias1.append(gene[index])
      index+=1
    self.weights2 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      self.weights2.append(gene[index])
      index+=1
    self.bias2=gene[index]

  def getContents(self):
    gene=[]
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      for j in range(0,col-1): 
        gene.append(self.weights1[i][j])
  
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      gene.append(self.bias1[i])
  
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      gene.append(self.weights2[i])

    gene.append(self.bias2)

    return gene

  def setrandom(self):
    self.weights1 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      arr = []
      for j in range(0,col-1): 
        arr.append(float(random.uniform(-1.0,1.0)))
      self.weights1.append(arr)
    self.bias1 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      inp = 0
      while(inp == 0):
        inp = float(random.uniform(-1.0,1.0))
        self.bias1.append(inp)
    self.weights2 = []
    for i in range (0,hiddenNodes):# hiddenNodes nodes
      self.weights2.append(float(random.uniform(-1.0,1.0)))
    inp = 0
    while(inp == 0):
      inp = float(random.uniform(-1.0,1.0))
      self.bias2 = inp


  def calcFitness(self):
    yhat=[]
    for r in range (0,row):
      As = []
      #iterate through rows of weights1
      for i in range (0,hiddenNodes):# hiddenNodes nodes
        z1=0
        # iterate through rows of Is(dataset.loc[r])
        for k in range(0,4):
          z1 = z1 + (self.weights1[i][k] * dataset.loc[r][k])
        z1 += self.bias1[i]
        z1 = 1/(1+math.e**(-z1)) #sigmoid
        As.append(z1)

      z2 = 0
      for i in range (0,hiddenNodes):# hiddenNodes nodes
        z2 = z2 + (As[i] * self.weights2[i])
      z2 += self.bias2
      z2 = 1/(1+math.e**(-z2)) #sigmoid
      #print(r," : ",z2)
      if(z2 > 0.5):
        yhat.append(1)
      else:
        yhat.append(0)

    count=0
    for i in range (0,row):
      if(dataset.loc[i][4] == yhat[i]):
        count = count+1

    Accuracy = (count/row)*100
    if(Accuracy >= 95):
      self.setrandom()
      self.calcFitness()
    else:
      self.fitness=Accuracy


  def printt(self):
    print('----------------------------------------------------------')


class ANN:
  weightVectorList=[]
  popSize=20

  def __init__(self):

    for i in range(self.popSize):
      newweightVector = weightVector()
      self.weightVectorList.append(newweightVector)

  def geneticSolution(self):
    for generation_count in range(1000):
      rankedweightVectors=[]
      for i in range(self.popSize):
        self.weightVectorList[i].calcFitness()
        rankedweightVectors.append((self.weightVectorList[i]))

      rankedweightVectors.sort(key=lambda x: x.fitness, reverse=True)

      print(f"=== Generation {generation_count} best solution ====")
      print(rankedweightVectors[0].fitness)
      print(" ")
      rankedweightVectors[0].printt()
      print(" ")

      if rankedweightVectors[0].fitness >=85:
        dataset2 = pd.read_csv('iris_dataset_testing.csv')
        row2 = dataset2.shape[0]
        col2 = dataset2.shape[1]
        yhat=[]
        for r in range (0,row2):
          As = []
          #iterate through rows of weights1
          for i in range (0,hiddenNodes):# hiddenNodes nodes
            z1=0
            # iterate through rows of Is(dataset.loc[r])
            for k in range(0,4):
              z1 = z1 + (rankedweightVectors[0].weights1[i][k] * dataset2.loc[r][k])
            z1 += rankedweightVectors[0].bias1[i]
            z1 = 1/(1+math.e**(-z1)) #sigmoid
            As.append(z1)

          z2 = 0
          for i in range (0,hiddenNodes):# hiddenNodes nodes
            z2 = z2 + (As[i] * rankedweightVectors[0].weights2[i])
          z2 += rankedweightVectors[0].bias2
          z2 = 1/(1+math.e**(-z2)) #sigmoid
          if(z2 > 0.5):
            yhat.append(1)
          else:
            yhat.append(0)

        count=0
        for i in range (0,row2):
          if(dataset2.loc[i][4] == yhat[i]):
            count = count+1
        Accuracy = (count/row)*100
        print("TESTING ACCURACY: ",Accuracy," %")
        break

      newANN=[]

      for i in range(int(self.popSize*(20/100))):
        newweightVector = weightVector()
        newweightVector.setContents(rankedweightVectors[i].getContents())
        newANN.append(newweightVector)


      for j in range(int(self.popSize*(40/100))):
        first = int(random.uniform(0,self.popSize*(80/100)))
        second = int(random.uniform(0,self.popSize*(80/100)))
        pivot = int(random.uniform(0,hiddenNodes*6)) #total w+b
        array = []
        contents1=rankedweightVectors[first].getContents()
        contents2=rankedweightVectors[second].getContents()
        for i in range (0,pivot):
          array.append(contents1[i])
        for i in range (pivot,hiddenNodes*6+1):
          array.append(contents2[i])
        newweightVector = weightVector()
        newweightVector.setContents(array)
        newANN.append(newweightVector)

      #mutation
      for i in range(int(self.popSize*(40/100))):
        k = int(random.uniform(0,self.popSize*(80/100)))
        array=rankedweightVectors[k].getContents()
        for j in range(0,int(hiddenNodes*6/4)):
          index=int(random.uniform(0,hiddenNodes*6))
          array[index]=float(random.uniform(-1.0,1.0))
        newweightVector = weightVector()
        newweightVector.setContents(array)
        newANN.append(newweightVector)

      self.weightVectorList=[]
      self.weightVectorList= newANN
      

solution = ANN()
ANN.geneticSolution(solution)
```
